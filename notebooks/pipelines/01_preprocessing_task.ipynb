```python
import sys
import os
import yaml
import mne
import matplotlib.pyplot as plt

project_root = os.path.abspath('../../')
if project_root not in sys.path:
    sys.path.append(project_root)

from src.data.discovery import discover_task_files
from src.data.loader import load_task_epochs
from src.preprocessing.core import apply_feature_preprocessing
```

```python
with open('../../configs/experiments/data_config.yaml', 'r') as f:
    data_cfg = yaml.safe_load(f)
with open('../../configs/preprocessing/task_feature_hmm.yaml', 'r') as f:
    pre_cfg = yaml.safe_load(f)

# Simulamos cargar a EDGSSINGO (o cualquier otro sujeto)
df = discover_task_files(data_cfg['dataset']['root_dir'])
sample_subject = df.iloc[0] # Cambia el índice si quieres a alguien en específico
epochs_raw = load_task_epochs(sample_subject['ruta'], data_cfg['dataset']['sampling_rate'], data_cfg['channels']['eeg'])
```

```python
print("--- ANTES DEL PREPROCESAMIENTO ---")
epochs_raw.plot_psd(fmin=1, fmax=50, average=True, spatial_colors=False)

epochs_clean = apply_feature_preprocessing(epochs_raw, pre_cfg)

print("\n--- DESPUÉS DEL PREPROCESAMIENTO (1-30 Hz + CAR) ---")
epochs_clean.plot_psd(fmin=1, fmax=50, average=True, spatial_colors=False)

# Imprimir el reporte de épocas tiradas por la borda
print(f"\nÉpocas originales: {len(epochs_raw)}")
print(f"Épocas limpias supervivientes: {len(epochs_clean)}")
print(f"Porcentaje de retención: {(len(epochs_clean)/len(epochs_raw))*100}%")
```
